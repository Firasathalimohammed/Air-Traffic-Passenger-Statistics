### 1. What is the trend in passenger traffic over time, and are there any seasonal patterns in air travel activity?

```python
df['Activity Period Start Date'] = pd.to_datetime(df['Activity Period Start Date'])

passenger_traffic = df.groupby('Activity Period Start Date')['Passenger Count'].sum()

plt.figure(figsize=(12, 6))
plt.plot(passenger_traffic.index, passenger_traffic.values, marker='o', linestyle='-', color='b')
plt.title('Passenger Traffic Trend Over Time')
plt.xlabel('Activity Period')
plt.ylabel('Total Passenger Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

1. <b>Overall Increasing Trend</b>: Over the years from 1999 to 2023, there is a noticeable upward trend in <b>passenger traffic</b>. The total <b>passenger count</b> has generally increased, which indicates a growing demand for <b>air travel</b> over time.<br>
<br>
2. <b>Seasonal Fluctuations</b>: The data exhibits <b>seasonal patterns</b>. There are periodic fluctuations in <b>passenger traffic</b>, with peaks during certain months and lower counts during others. These patterns may be attributed to factors such as holidays, vacations, or travel seasons.<br>
<br>
3. <b>Recent Variations</b>: In recent years, especially in 2023, there seems to be a fluctuation in <b>passenger traffic</b>, with both increases and decreases in counts. This suggests that external factors like economic conditions, travel restrictions, or global events may impact <b>air travel</b> trends.<br>

### 2. How do different airlines compare in terms of passenger counts, and are there any significant variations between domestic and international flights?

```python
passenger_counts = df.groupby(['Operating Airline', 'GEO Summary'])['Passenger Count'].sum().unstack()

top_5_domestic = passenger_counts['Domestic'].nlargest(5)
top_5_international = passenger_counts['International'].nlargest(5)

plt.figure(figsize=(12, 8))
top_5_domestic.plot(kind='bar', color='b', alpha=0.6, label='Domestic')
top_5_international.plot(kind='bar', color='g', alpha=0.6, label='International')
plt.title('Top 5 Airlines by Passenger Counts (Domestic vs. International)')
plt.xlabel('Operating Airline')
plt.ylabel('Total Passenger Count')
plt.legend(title='GEO Summary', title_fontsize='12')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

* The airline with the highest passenger count for domestic flights is "<b>United Airlines</b>" with a total of 166,103,646 passengers.<br>
<br>
* The second-ranking airline for domestic flights is "<b>United Airlines - Pre 07/01/2013</b>" with 159,578,077 passengers.
"* <b>American Airlines</b>" is in third place with 74,729,337 passengers.<br>
<br>
"<b>SkyWest Airlines</b>" and "<b>Delta Air Lines</b>" occupy the fourth and fifth positions with 68,955,624 and 61,063,369 passengers, respectively.<br>
<br>
* "<b>United Airlines - Pre 07/01/2013</b>" is the leading airline for international flights with 40,042,459 passengers.<br>
<br>
* "<b>United Airlines</b>" takes the second position with 34,489,889 passengers.<br>
<br>
* "<b>Air Canada</b>" ranks third with 15,126,884 passengers.<br>
<br>
* "<b>Lufthansa German Airlines</b>" and "<b>British Airways</b>" secure the fourth and fifth spots with 10,055,167 and 9,368,317 passengers, respectively.<br>

### 3. Are there specific geographic regions or airports that show a notable increase in passenger traffic, and what factors might contribute to these variations?

```python
passenger_counts_by_region = df.groupby('GEO Region')['Passenger Count'].sum()

plt.figure(figsize=(8, 8))
plt.pie(passenger_counts_by_region, labels=passenger_counts_by_region.index, autopct='%1.f%%', startangle=140,
       textprops={"size":10,"fontweight":"black"})
plt.title('Passenger Counts by Geographic Region (Pie Chart)')
plt.axis('equal')
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

* The <b>US</b> is the dominant contributor to passenger traffic among all the <b>GEO Regions</b>, with a significantly higher total of <b>747,120,929</b> passengers. This suggests the substantial role the United States plays in air travel.<br>
<br>
* Among the other regions, <b>Asia</b> and <b>Europe</b> stand out as significant contributors to passenger counts, with <b>92,685,649</b> and <b>62,543,779</b> passengers, respectively. These regions also play a substantial role in international air travel.<br>
<br>
* In contrast, <b>South America</b> registers a comparatively lower passenger count of <b>250,822</b>. It indicates that air travel to and from South America may be less prominent compared to the other regions listed.<br>

### 4. What is the relationship between airfare price categories and passenger counts, and does low-fare pricing significantly impact passenger numbers?

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price Category Code', y='Passenger Count', data=df, palette='Set2')
plt.title('Relationship between Airfare Price Categories and Passenger Counts')
plt.xlabel('Price Category Code')
plt.ylabel('Passenger Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="lefAt"><font color=brown>üí¨ Inference:</font></h3>

1. Variability Across Price Categories: The box plot reveals that there is considerable variability in passenger counts across different airfare <b>price categories</b>. The boxes represent the <b>interquartile range</b>, and the whiskers show the spread of data, indicating that some <b>price categories</b> have a wider range of passenger counts than others.<br>
<br>
    
2. Outliers and Extreme Values: The presence of <b>outliers</b> in the box plot indicates instances where <b>passenger counts</b> significantly deviate from the norm. These <b>outliers</b> might correspond to extraordinary circumstances or anomalies that need further investigation to understand their impact on <b>passenger numbers</b>.<br>
